```python
import pandas as pd
from scipy.stats import skew, kurtosis

data = pd.read_csv('data.csv')
```

```python
data
```

```python
#To pick standard deviation of a column
data.describe()
```

```python
# Tp groupby a particular column and then check the smallest fuel cost per unit burned for each fuel type

data.groupby('fuel_type_code_pudl')['fuel_cost_per_unit_burned'].mean().reset_index().sort_values('fuel_cost_per_unit_burned')
```

```python
#To check for the skew of a column, put the column in the finction skew()

print(skew(data['fuel_qty_burned']))
print(kurtosis(data['fuel_qty_burned']))
```

```python
# To check for missing value
data.isnull().sum()
```

```python
# To check for jow all the columns correlate with column 'fuel_cost_per_unit_burned'

data.corr()['fuel_cost_per_unit_burned'].reset_index().sort_values('fuel_cost_per_unit_burned')
```

```python
#To check for highest fuel cost delivered using groupby

groupby = data.groupby('report_year')['fuel_cost_per_unit_delivered'].mean().reset_index()
groupby.sort_values('fuel_cost_per_unit_delivered', ascending = False)
```

```python
#To check the difference between fuel cost per unit burned in two years for a given fuel type
# For instance to check for coal in year 1994 and 1998 will be
# (14984.572 - 11902.597)/14984.572

data.groupby(['fuel_type_code_pudl','report_year'])['fuel_cost_per_unit_burned'].sum().reset_index()
```
